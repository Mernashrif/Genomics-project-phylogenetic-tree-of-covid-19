## format preferences

Here is where I just took code from Jesse's notebook and calculated the "probabilities" for expression and for binding.

### calculate probabilities

```python
mut_effects = pd.read_csv('data/single_mut_effects.csv')
```

```python
# this section is from Jesse's notebook
conditions = ['bind', 'expr']
# get mutation effects without stop codons
nostop_mut_effects = mut_effects.query('(mutant != "*") and (wildtype != "*")')

# exponent scaling factors alpha for binding and expression
manual_scale_alpha = 1.4  # ad hoc additional scaling factor for exponents
alpha_bind = math.log(10) * manual_scale_alpha
alpha_expr = alpha_bind * ((nostop_mut_effects['expr_avg'].max() - nostop_mut_effects['expr_avg'].min()) /
                           (nostop_mut_effects['bind_avg'].max() - nostop_mut_effects['bind_avg'].min()))
print(f"alpha (exponent scale factor):\nbinding = {alpha_bind:.3f}\nexpression = {alpha_expr:.3f}")

def compute_site_info(p):
    p = p[p > 0]
    assert len(p), 'no non-zero values'
    return math.log(20) + sum(p * np.log(p))

df = (
    nostop_mut_effects
    .assign(nonorm_prob_bind=lambda x: np.exp(alpha_bind * x['bind_avg']).fillna(0),
            nonorm_prob_expr=lambda x: np.exp(alpha_expr * x['expr_avg']).fillna(0),
            sitesum_prob_bind=lambda x: x.groupby('site_RBD')['nonorm_prob_bind'].transform('sum'),
            sitesum_prob_expr=lambda x: x.groupby('site_RBD')['nonorm_prob_expr'].transform('sum'),
            prob_bind=lambda x: x['nonorm_prob_bind'] / x['sitesum_prob_bind'],
            prob_expr=lambda x: x['nonorm_prob_expr'] / x['sitesum_prob_expr']
            )
    [['site_SARS2', 'site_RBD', 'wildtype', 'mutant',
      'prob_bind', 'prob_expr', 'bind_avg', 'expr_avg']]
    ).rename(columns={'site_RBD': 'site'})

assert not df[['prob_bind', 'prob_expr']].isnull().any(axis=None)

df.head().round(3)
```

### create various preference sets

```python
# each preference by itself
prefs = {}
for condition in conditions:
    temp = df.pivot(index='site', columns='mutant', values=f'prob_{condition}')
    assert np.allclose(temp.sum(axis=1), 1, atol=0.005)
    prefs[condition] = temp
    temp.reset_index().to_csv(f'results/prefs/{condition}.csv', index=False)
```

```python
# preference set is the two multiplied together
condition = 'expr_times_bind'
conditions.append(condition)
df[condition] = df['prob_bind'] * df['prob_expr']
temp = df.pivot(index='site', columns='mutant', values=condition)
temp = temp.div(temp.sum(axis=1), axis=0)
assert np.allclose(temp.sum(axis=1), 1, atol=0.005)
prefs[condition] = temp
temp.reset_index().to_csv(f'results/prefs/{condition}.csv', index=False)
```

```python
# preference set is the more constrained profile of the two
condition = 'max_entropy'
conditions.append(condition)
temp = []
bind = prefs['bind']
expr = prefs['expr']
assert set(bind.index.values) == set(expr.index.values)
bind_entropies = bind.apply(entropy, axis=1).tolist()
expr_entropies = expr.apply(entropy, axis=1).tolist()
for i in range(len(bind_entropies)):
    if bind_entropies[i] < expr_entropies[i]:
        temp.append(i+1)
    elif bind_entropies[i] > expr_entropies[i]:
        pass
    else:
        print('tie!')
temp = pd.concat([bind.loc[temp], expr.loc[[x for x in bind.index.values if x not in temp]]]).sort_index()
assert len(temp) == len(bind) == len(expr)
prefs[condition] = temp
temp.reset_index().to_csv(f'results/prefs/{condition}.csv', index=False)
```

```python
prefs = " ".join([f'results/prefs/{condition}.csv' for condition in conditions])
prefs
```

### rescale prefs

```python
rescaled_prefs = []
for model in expcm_models:
    condition = model.split("_")[1]
    beta = df[df.Model == model]['beta'].iloc[0]
    temp = prefs[condition] ** beta
    temp = temp.div(temp.sum(axis=1), axis=0)
    assert np.allclose(temp.sum(axis=1), 1, atol=0.005)
    temp = temp.reset_index()
    temp = pd.melt(temp, id_vars='site', var_name='mutation', value_name='mut_rescaled prefs')
    temp['condition'] = model
    rescaled_prefs.append(temp)
rescaled_prefs = pd.concat(rescaled_prefs)
rescaled_prefs
```
